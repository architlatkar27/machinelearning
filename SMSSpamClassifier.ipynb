```python
rawData = open("SMSSpamCollection.tsv").read()
```

```python
rawData[:500]
```

```python
parseData = rawData.replace('\t','\n').split('\n')
```

```python
parseData[:4]
```

```python
labelList = parseData[::2]
smsList = parseData[1::2]
```

```python
import pandas as pd
import numpy as np

message_set = pd.DataFrame({
    'label': labelList[:-1],
    'message': smsList    
})
```

```python
message_set.head()
```

```python
message_set.shape
```

```python
message_set['label'].value_counts()
```

```python
import re
word_list = []
for m in message_set['message']:
    wlist = re.findall('\w+',m)
    word_list.append(wlist)
message_set['message_break'] = word_list
message_set.head()
```

```python
import string
def rem_punc(text):
    new_text = "".join(char for char in text if char not in string.punctuation)
    return new_text

message_set['text_no_punc'] = message_set['message'].apply(lambda x:rem_punc(x))
message_set.head()
```

```python
del message_set['message_break']
```

```python
tokenize = []
for m in message_set['text_no_punc']:
    l = re.split('\W+',m)
    for i in range(len(l)):
        l[i] = l[i].lower()
    tokenize.append(l)
message_set['token_set'] = tokenize
message_set.head()
```

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')
print(stopwords)
```

```python
nostop = []
for l in message_set['token_set']:
    new_l = list(filter(lambda x: x not in stopwords,l))
    nostop.append(new_l)
message_set['text_nostop'] = nostop
message_set.head()
```

```python
ps = nltk.PorterStemmer()
```

```python
stem = []
for l in message_set['text_nostop']:
    new_l = []
    for w in l:
        new_l.append(ps.stem(w))
    stem.append(new_l)
message_set['text_nostem'] =  stem
message_set.head()
```

```python
wnl = nltk.WordNetLemmatizer()
```

```python
lemma = []
for l in message_set['text_nostop']:
    new_l = []
    for w in l:
        new_l.append(wnl.lemmatize(w))
    lemma.append(new_l)
message_set['text_lemmatized'] = lemma
message_set.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cvec = CountVectorizer(analyzer='word')
x_counts = cvec.fit_transform(message_set['message'])
print(x_counts)
vec_dataset = pd.DataFrame(x_counts.toarray())
vec_dataset
```

```python
from sklearn.preprocessing import LabelEncoder
le_y = LabelEncoder()
y = message_set.iloc[:,0].values
y = le_y.fit_transform(y)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
X = x_counts.toarray()
scores = cross_val_score(clf,X,y,cv=10)
print(scores)
print(scores.mean())
```
